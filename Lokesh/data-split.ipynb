```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
train = pd.read_csv('train.csv')
```

```python
train.shape
```

```python
train
```

```python
t1 = train[train['Target']==1]
t0 = train[train['Target']==0]
```

```python
train_percent = 70
test_percent = 100 - train_percent

trt1 = t1.sample(len(t1)*train_percent//100)
tet1 = t1.drop(trt1.index)
trt0 = t0.sample(len(t0)*train_percent//100)
tet0 = t0.drop(trt0.index)

train_final = pd.concat([trt1,trt0])
test_final = pd.concat([tet1,tet0])
```

```python
train_final = train_final.drop(['Primary key'],axis=1)
test_final = test_final.drop(['Primary key'],axis=1)
```

```python
train_final.to_csv('train_final.csv',index=False)
test_final.to_csv('test_final.csv',index=False)
```
